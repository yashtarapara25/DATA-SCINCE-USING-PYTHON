```python
sns.histplot(dimonds['price'])
plt.show()
```

```python
sns.histplot(dimonds['price'],kde=True)
plt.show()
```

```python
sns.distplot(dimonds['price'],color='purple',kde=False,bins=10)
plt.show()
```

```python
sns.distplot(dimonds['price'],color='purple',kde=False,bins=10,vertical=True)
plt.show()
```
